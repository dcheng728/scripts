# Indefinite Integral

```python
from sage.symbolic.integration.integral import indefinite_integral
```

```python
var('x m k T')
```

```python
f(x) = x^2 * e^(-m*x^2/(2*k*T))
```

```python
view(indefinite_integral(f(x), x))
```

# Definite Integral

```python
from sage.symbolic.integration.integral import definite_integral
```

```python
var('x m k T')
```

```python
var('pp p x L')
```

```python
L = 10
```

```python
f(x) = e^(i*(pp)*x)
```

```python
res = f.integral(x,0,L)
```

```python
view(res)
```

```python
res(pp = 0.000001)
```

```python
print(res(pp = 0.1).abs())
```

```python
for i in range(1,1000):
    print(0.1 * i,res(pp = 0.001 * i).abs())
```

```python
plot(res,pp)
```
