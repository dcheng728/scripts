```python
import torch
import matplotlib.pyplot as plt
```

```python
def find_best_f(n):
    a = torch.nn.Parameter(torch.rand(n),requires_grad = True)
    op = torch.optim.Adam([a], lr=0.005)
    for i in range(10000):
        op.zero_grad()

        c1 = (torch.sum(a))**2 
        c2 = (torch.sum(a**2)-1)**2
        f = torch.sum(a * torch.roll(a,-1))

        l = 1000000 * (c1 + c2) - f

        l.backward()
        op.step()
        
    #print(c1.detach().numpy(),c2.detach().numpy(),f.detach().numpy(),l.detach().numpy())
    
    return f
```

```python
Is = []
Fs = []

for i in range(2,10):
    print(i)
    Is.append(i)
    Fs.append(find_best_f(i).detach().numpy())
```

```python
plt.plot(Is,Fs)
```
